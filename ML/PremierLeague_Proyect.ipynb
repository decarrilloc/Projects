```python
years=list(range(2023,2021,-1))
all_matches=[]
standings_url='https://fbref.com/en/comps/9/Premier-League-Stats'
```

```python
table_2
```

```python
for year in years:
    data=requests.get(standings_url)
    soup=BeautifulSoup(data.text)
    regular_season=soup.select('table.stats_table')[0]
    links=regular_season.find_all('a')
    links=[l.get('href') for l in links]
    links=[l for l in links if 'squads' in l]
    team_urls = [f"https://fbref.com{l}" for l in links]
    
    prev_season_url=soup.select('div.prevnext a[href]')[0].get('href')
    standings_url=f'https://fbref.com{prev_season_url}'

    for team_url in team_urls:
        team_name=team_url.split('/')[-1].replace('-Stats','').replace('-','')
        data=requests.get(team_url)
        matches=pd.read_html(data.text,match='Scores & Fixtures')[0]
        soup=BeautifulSoup(data.text)
        soup_1=soup.find_all('a')
        url=[l.get('href') for l in soup_1]
        links=[l for l in url if l and 'all_comps/shooting/' in l]
        links=[f'https://fbref.com{l}' for l in links]
        table_1_url=requests.get(links[0])
        shooting=pd.read_html(table_1_url.text,match='Shooting')[0]
        shooting.columns = shooting.columns.droplevel()
        table_2=[l for l in url if l and 'all_comps/gca/' in l]
        table_2=f'https://fbref.com{table_2[0]}'
        table_2_url=requests.get(table_2)
        goal_and_shoot=pd.read_html(table_2_url.text,match='Goal and Shot Creation')[0]
        goal_and_shoot.columns=goal_and_shoot.columns.droplevel()
        table_3=[l for l in url if l and 'all_comps/passing/' in l]
        table_3=f'https://fbref.com/{table_3[0]}'
        table_3_url=requests.get(table_3)
        passing=pd.read_html(table_3_url.text,match='Passing')[0]
        passing.columns=passing.columns.droplevel()
        table_4=[l for l in url if l and 'all_comps/defense/' in l]
        table_4=f'https://fbref.com/{table_4[0]}'
        table_4_url=requests.get(table_4)
        defensive_act=pd.read_html(table_4_url.text,match='Defensive Actions')[0]
        defensive_act.columns=defensive_act.columns.droplevel()

        
        
        try:
            team_data = matches.merge(shooting[['Date','Gls', 'Sh', 'SoT', 'SoT%', 'G/Sh', 'G/SoT', 'Dist', 'FK','PK', 'PKatt', 'xG', 'npxG', 'npxG/Sh', 'G-xG', 'np:G-xG']],on='Date')
            team_data=team_data.merge(passing[['Date', 'Cmp', 'Att', 'Cmp%', 'TotDist', 'PrgDist','Ast', 'xAG', 'xA','KP', '1/3', 'PPA', 'CrsPA', 'PrgP']],on='Date')
            team_data=team_data.merge(goal_and_shoot[['Date','SCA', 'PassLive', 'PassDead', 'TO', 'Fld', 'Def','GCA', 'PassLive', 'PassDead', 'TO', 'Sh', 'Fld', 'Def']],on='Date')
            team_data=team_data.merge(defensive_act[['Date','Tkl', 'TklW', 'Def 3rd', 'Mid 3rd', 'Att 3rd', 'Tkl','Att', 'Tkl%', 'Lost', 'Blocks', 'Sh', 'Pass', 'Int', 'Tkl+Int', 'Clr','Err']],on='Date')
        
        except ValueError:
            continue
        
        team_data["Season"] = year
        team_data["Team"] = team_name
        all_matches.append(team_data)
        time.sleep(5)
```

```python
match_df = pd.concat(all_matches)
match_df
```
